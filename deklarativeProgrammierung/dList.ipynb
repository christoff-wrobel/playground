```python
# Type dList:
# A dList is a simple linked list allowing declarative programming. The two central data structures are dLists consisting of an 
# element (car) and a dList (cdr). 
# car: the head of the list
# cdr: the rest of the list
# To identify the end of a list a data structure called emptyDList is used.
# Example: The python list [1,2,3] is mapped in symbols to a dList 
# 1 -> 2 -> 3 -> emptyDList
# The abstract base class dListTool implements all usually needed methods to work with simple linked lists.

class emptyDList:
    pass
    
class dList:
    car = None
    cdr = None

    def __init__(self, aCar, aCdr):
        self.car = aCar
        self.cdr = aCdr



from abc import ABC
class dListTool(ABC):
# Konstruktoren
    def makeEmptyDList():
        return emptyDList()
    
    def cons(e,l=emptyDList()):
        return dList(e,l)

    def makeDList(l):
        ret = dListTool.makeEmptyDList()
        for x in l:
            ret = dListTool.cons(x,ret)
        return dListTool.reverse(ret)
    
        

    
# Praedikate
    def isEmptyDList(l):
        return isinstance(l,emptyDList)

    def isDList(l):
        return isinstance(l,dList)


    


    
# Ausgabe
    def print(l):
        if dListTool.isEmptyDList(l):
            return
        else:
            print(l.car)
            dListTool.print(l.cdr)


# Weitere Methoden
    def len(l, akk=0):
        if dListTool.isEmptyDList(l):
            return akk
        else:
            return dListTool.len(l.cdr, akk + 1)

    def reverse(l, temp= None):
        if temp == None:
                temp = dListTool.makeEmptyDList()
        if dListTool.isEmptyDList(l):
            return temp
        else:
            return dListTool.reverse(l.cdr,
                                     dListTool.cons(l.car,
                                                    temp))



    def append(l,m, temp=None):
        if temp == None:
            temp = dListTool.reverse(l)
        if dListTool.isEmptyDList(m):
            return dListTool.reverse(temp)
        else:
            return dListTool.append(l,m.cdr, dListTool.cons(m.car,temp))

    
    def getEntry(l, pos):
        if dListTool.isEmptyDList(l):
            print("!!! Error in method getEntry !!!")
            return None   
        if pos == 0:
            return l.car
        else:
            return dListTool.getEntry(l.cdr,pos-1)
            
    def setEntry(e, l, pos, temp = None):
        if dListTool.isEmptyDList(l):
            print("!!! Error in method setEntry !!!")
            return None  
            
        if temp == None:
            temp = emptyDList()
         
        if pos == 0:
            return dListTool.append(dListTool.reverse(temp), dListTool.cons(e, l))
            
        else:
            return dListTool.setEntry(e, 
                                      l.cdr,
                                      pos-1, 
                                      dListTool.cons(l.car, temp))



#Example
    def run():
        print(isinstance([1,2,3],list))
        print(type([1,2,3]))
        
        print()
        x=dListTool.cons(2,dListTool.cons(1,dListTool.makeEmptyDList()))
        dListTool.print(x)
        
        print()
        y = dListTool.makeDList([1,2,3,4,5])
        dListTool.print(y)
        
        print()
        print(dListTool.len(y))
        
        print()
        print(dListTool.getEntry(y,0))
        print(dListTool.getEntry(y,1))
        
        print()
        dListTool.print(dListTool.reverse(y))

        print()
        print("Append")
        dListTool.print(x)
        dListTool.print(y)
        dListTool.print(dListTool.append(x,y))
        
        print()
        print("setEntry")
        dListTool.print(dListTool.setEntry('a',y,3))



dListTool.run()
    
# todos
    # setEntry
    # removeEntry
    # append
    # filter
    # sort
    # map
    # fold
    
        
            


    
```
